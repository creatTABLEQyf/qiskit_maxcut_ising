```python
# no periodic boundary
df = pd.read_excel('5x5.xlsx')
seeds = df['seed']
Energy = df['energy'].tolist()
iterations = df['terminate iterations'].tolist()
execution_time = df['time (s)'].tolist()
```

```python
# seed v.s. execution time
plt.figure(dpi = 300)
plt.plot(seeds, np.array(execution_time)/3600, 'o')
plt.xlabel('Seed')
plt.ylabel('Execution time (h)')
plt.title('5x5')
plt.grid()
plt.show()
```

```python
# seed v.s. iterations
plt.figure(dpi = 300)
plt.plot(seeds, iterations, 'go')
plt.xlabel('Seed')
plt.ylabel('Terminate Iterations')
plt.title('5x5')
plt.grid()
plt.show()
```

```python
# seed v.s. iterations
plt.figure(dpi = 300)
plt.plot(iterations, execution_time, 'mo')
plt.xlabel('Terminate Iterations')
plt.ylabel('Execution time (s)')
plt.title('5x5')
plt.grid()
plt.show()
```

```python
# seeds v.s. energy
plt.figure(dpi = 300)
plt.plot(seeds, Energy, 'ro')
plt.xlabel('Seed')
plt.ylabel('Final Energy')
plt.title('5x5')
plt.grid()
plt.show()
```

```python
# Cobyla(maxiter = 400)
seeds = np.linspace(201, 10000, 10000 - 200)
Energy = []
iterations = []
execution_time = []
```

```python
# Cobyla
df = pd.read_excel('COBYLA(2000)_D1.xlsx')
Energy += df['energy'].tolist()
iterations += df['terminate iterations'].tolist()
execution_time += df['time (s)'].tolist()

# seed v.s. execution time
plt.figure(dpi = 600)
plt.plot(seeds, execution_time, '.')
plt.xlabel('Seed')
plt.ylabel('Execution time (s)')
plt.title('COBYLA(maxiter = 2000)')
plt.grid()
plt.show()
```

```python
# seed v.s. iterations
plt.figure(dpi = 300)
plt.plot(seeds, iterations, 'g.')
plt.xlabel('Seed')
plt.ylabel('Terminate Iterations')
plt.title('COBYLA(maxiter = 2000)')
plt.grid()
plt.show()
```

```python
# seed v.s. iterations
plt.figure(dpi = 300)
plt.plot(iterations, execution_time, 'm.')
plt.xlabel('Terminate Iterations')
plt.ylabel('Execution time (s)')
plt.title('COBYLA(maxiter = 2000)')
plt.grid()
plt.show()
```

```python
# seeds v.s. energy
plt.figure(dpi = 300)
plt.plot(seeds, Energy, 'r.')
plt.xlabel('Seed')
plt.ylabel('Ground State Energy')
plt.title('COBYLA(maxiter = 2000)')
plt.grid()
plt.show()
```
